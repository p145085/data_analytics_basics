```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('interstellar_travel.csv')
```

```python
df
```

```python
df[df['Transportation Type'] == 'Warp Drive']['Destination'].drop_duplicates().dropna()
# The destinations which were travelled to using 'Warp Drive'. 
# .dropna() because we only care about the known destinations although NaN implies we travelled somewhere else.
```

```python
df[df['Transportation Type'] == 'Ion Thruster']['Star System'].drop_duplicates().dropna()
# Similar to previous cell, this time we're looking at Ion Thrusters and what 'Star System' the destination is in.
```

```python
df['Duration of Stay (Earth Days)'].max()
# The longest duration of stay anywhere. 14 months and then some.
```

```python
df
```

```python
df['Most Expensive Stay'] = df['Price (Galactic Credits)'] / df['Duration of Stay (Earth Days)']

df['Most Expensive Stay'].replace(-np.inf, np.nan, inplace=True)
df['Most Expensive Stay'].replace(np.inf, np.nan, inplace=True)

sorted_df = df.sort_values(by='Most Expensive Stay')

sorted_df['Most Expensive Stay'].dropna()

# -1481.287500 -> Indicating profit was made on this trip.
# 12076.015259 -> Most expensive stay.

first_index = sorted_df['Most Expensive Stay'].idxmin()
last_index = sorted_df['Most Expensive Stay'].idxmax()

first_expensive_row = sorted_df.loc[first_index]
last_expensive_row = sorted_df.loc[last_index]

# Display the entire rows
print("Least expensive stay:")
print(first_expensive_row)
print("\n")
print("Most expensive stay:")
print(last_expensive_row)
```

```python
df['Booking Date'] = pd.to_datetime(df['Booking Date'])
df['Departure Date'] = pd.to_datetime(df['Departure Date'])

df['Time from Booking to Departure'] = df['Departure Date'] - df['Booking Date']

# New column made.

df.sort_values(by = 'Time from Booking to Departure')['Time from Booking to Departure']

idx = df['Time from Booking to Departure'].idxmax()
val = df['Time from Booking to Departure'][idx]
res = val / pd.Timedelta(days=365)
res

# There seems to be a maximum of 4 years in advance booking.
```

```python
cross_tab = pd.crosstab(df['Occupation'], df['Transportation Type'])

# Plotting
plt.figure(figsize=(10, 6))
sns.heatmap(cross_tab, annot=True, cmap='coolwarm', fmt='d')
plt.title('Correlation between Occupation and Transportation Type')
plt.xlabel('Transportation Type')
plt.ylabel('Occupation')
plt.show()

# We learn that tourists, the supposedly less educated group uses the most cost-efficient mode of travel, solar sailing.
# Scientists steer away from using not-well-established modes of travel (other). Although they ought to also be the ones pioneering new modes.
# Tourists also have an uptick in using "other" modes of transportation, possibly some exploitation of not knowing better.
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Distance to Destination (Light-Years)', y='Number of Companions')
plt.title('Correlation between Distance to Destination (Light-Years) and Number of Companions')
plt.xlabel('Distance to Destination (Light-Years)')
plt.ylabel('Number of Companions')
plt.grid(True)
plt.show()

# We can deduce that there aren't really any cruises, ferries or haulers by this data.
# Interestingly the most amount of companions is part of the shortest distance travelled. 
# Furthermore the furthest distance travelled had but 1 companion. 
# We can assume these travels are made by specialized elite crews.
```

```python
correlation = df['Customer Satisfaction Score'].corr(df['Distance to Destination (Light-Years)'])

print("Correlation between Customer Satisfaction Score and Distance to Destination (Light-Years):", correlation)

# Negative 5 percent per distance travelled. Good number.
```

```python
from matplotlib.colors import LogNorm

pivot_table = df.pivot_table(index='Travel Class', columns='Age', aggfunc='size', fill_value=0)

plt.figure(figsize=(10, 6))
sns.heatmap(pivot_table, cmap='YlGnBu', annot=True, fmt='d', norm=LogNorm())
plt.title('Relationship between Travel Class and Age')
plt.xlabel('Age')
plt.ylabel('Travel Class')
plt.show()
```
